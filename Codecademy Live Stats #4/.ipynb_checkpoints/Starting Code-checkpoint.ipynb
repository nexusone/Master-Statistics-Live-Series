```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import random
```

```python
hourly_pay = np.genfromtxt('hourly_pay')
```

```python
hourly_pay
```

```python
plt.hist(hourly_pay, bins = 50)
```

```python
mean_salary = np.mean(hourly_pay)
np.round(mean_salary,2)
```

```python
population = list(hourly_pay)
samp1 = random.sample(population,150)
```

```python
np.mean(samp1)
```

```python
plt.hist(samp1, bins = 50)
plt.show()
```

```python
samp_means = []

for i in range (10000):
    samp = random.sample(population,150)
    samp_means.append(np.mean(samp))
```

```python
plt.hist(samp_means, bins = 50)
plt.xlim(0,40)
plt.show()
```

```python
np.mean(samp_means)
```

```python
from scipy import stats

#plot the sample means
sns.distplot(samp_means, bins = 50)
plt.xlim(0,40)

#plot the normal distribution
mu = np.mean(population)
sigma = np.std(population)/(150**.5)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x,stats.norm.pdf(x,mu,sigma), color='k')

plt.show()
```

```python
# standard error of the mean where 300 is the sample size
# standard error is a description of how wide the distribution is
# standard deviation divided by square root of sample size
# np.std(population)/(300**.5) 
```

```python
samp1_mean = np.mean(samp1)
samp1_std = np.std(samp1)
```

```python
#plot the sample means
sns.distplot(samp_means, bins = 50)
plt.xlim(0,40)

#plot the normal distribution
mu = samp1_mean
samp_size = len(samp1)
sigma = samp1_std/(samp_size**.5)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x,stats.norm.pdf(x,mu,sigma), color='k')

plt.show()
```
