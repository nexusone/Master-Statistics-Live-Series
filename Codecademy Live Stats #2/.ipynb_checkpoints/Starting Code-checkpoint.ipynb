```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import random
```

```python
rentals = pd.read_csv("https://raw.githubusercontent.com/Codecademy/datasets/master/streeteasy/streeteasy.csv")
```

```python
rentals.head()
```

```python
rentals.no_fee.replace({1:'yes', 0:'no'}, inplace=True)
rentals.has_roofdeck.replace({1:'yes', 0:'no'}, inplace=True)
rentals.has_washer_dryer.replace({1:'yes', 0:'no'}, inplace=True)
rentals.has_doorman.replace({1:'yes', 0:'no'}, inplace=True)
rentals.has_elevator.replace({1:'yes', 0:'no'}, inplace=True)
rentals.has_dishwasher.replace({1:'yes', 0:'no'}, inplace=True)
rentals.has_patio.replace({1:'yes', 0:'no'}, inplace=True)
rentals.has_gym.replace({1:'yes', 0:'no'}, inplace=True)
```

```python
rentals.describe(include='all')
```

## Univariate (one variable)

### Quantitative

```python
# Mean rent. sum rent divide by appartments
np.mean(rentals.rent)
rent_mean = rentals.rent.mean()
rent_mean
```

```python
# Mean trimmed. Meaning calculate the mean without the x (in this case 0.1 = 10%) values on min and max side, getting rid of outliers
from scipy import stats
rent_mean_trimmed = stats.trim_mean(rentals.rent, 0.1)
rent_mean_trimmed
```

```python
# Median rent. Sorted ascending, then take the middle value
rent_median = rentals.rent.median()
rent_median
```

```python
# Most common rent. Use with care on quantitive values
rent_mode = rentals.rent.mode()
rent_mode
```

```python
# plot the histogram with matplotlib, always uses 10 bins.
# add vlines to show the cutoff according to data
plt.hist(rentals.rent, bins = 30)
plt.vlines(x = rent_mean_trimmed, ymin = 0, ymax = 1000, colors='g', label = 'median')
plt.vlines(x = rent_mean, ymin = 0, ymax = 1000, colors='r', label='mean')
plt.legend()
plt.show()
```

```python
# plot the histogram with seaborn. creates bins according to data
sns.displot(rentals.min_to_subway)
```

```python
# what are those +40min subway appartments?
rentals[rentals.min_to_subway > 40].head()
```

```python
# only plot the appartments with less than 40 min subway times
sns.histplot(rentals[rentals.min_to_subway < 40].min_to_subway)
```
