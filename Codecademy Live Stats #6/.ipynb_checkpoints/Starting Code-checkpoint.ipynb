```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import random
from scipy import stats
```

```python
monthly_report = pd.read_csv('monthly_report.csv')
print(monthly_report.head())
```

|    **Null hypothesis:** | **is true**  | **is false**  |
|------------------------:|--------------|---------------|
| P-value below threshold | Type I Error | Correct!      |
| P-value above threshold | Correct!     | Type II error |

```python
print(np.sum(monthly_report.purchase == 'y'))
```

```python
# simulate 500 new visitor and calculate the p value
simulated_visitors = np.random.choice(['y','n'], size = 500, p=[.1,.9])
purchases = np.sum(simulated_visitors == 'y')
print(purchases)
p_value = stats.binom_test(purchases,500,.1)
print(p_value)
```

```python
# simulate 500 new visitors x times, run binomial test
null_purchases = []
p_values_sim = []
for i in range (10000):
    simulated_visitors = np.random.choice(['y','n'], size = 500, p=[.1,.9])
    purchases = np.sum(simulated_visitors == 'y')
    null_purchases.append(purchases)
    p_values_sim.append(stats.binom_test(purchases,500,.1))
```

```python
plt.hist(p_values_sim)
plt.show()
# uniform distribution
```

```python
plt.hist(null_purchases)
plt.show()
# normal distribution
```

|    **Null hypothesis:**                   | **is true**  | **is false**  |
|------------------------------------------:|--------------|---------------|
| P-value below threshold (significant)     | Type I Error | Correct!      |
| P-value above threshold (not significant) | Correct!     | Type II error |

```python
# simulate type 1 errors
type_one_errors = 0

for i in range (10000):
    simulated_visitors = np.random.choice(['y','n'], size = 500, p=[.1,.9])
    purchases = np.sum(simulated_visitors == 'y')
    p_value = stats.binom_test(purchases,500,.1)
    if p_value < 0.05:
        type_one_errors += 1
        
print(type_one_errors/10000)    
```
