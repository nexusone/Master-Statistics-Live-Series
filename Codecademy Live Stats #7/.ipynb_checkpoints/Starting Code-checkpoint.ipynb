# Two sample t-test
Null: There is no difference in mean resting blood pressure (among patients who do and don't have heart disease). 
True mean resting bp of ppl with hd - true mean resting bp of ppl without hd = 0

Alternative: There is a difference in mean resting blood pressure (among patients who do and don't have heart disease)
True mean resting bp of ppl with hd - true mean resting bp of ppl without hd != 0
